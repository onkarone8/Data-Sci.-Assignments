# Problem Statement.

PREDICT THE BURNED AREA OF FOREST FIRES WITH NEURAL NETWORKS.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report
import tensorflow as ts
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 300
%matplotlib inline
```

```python
df = pd.read_csv('forestfires.csv')
df
```

# Data Transformation

```python
df.drop(columns=['month','day'], inplace=True)
```

```python
df.head()
```

```python
sns.countplot(df['size_category'])
```

# Encoding Target Column

```python
le = LabelEncoder()
```

```python
df['size_category'] = le.fit_transform(df['size_category'])
```

# Feature Scaling

```python
x = df.iloc[:,:-1]
y = df['size_category']
```

```python
y
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)
x
```

# Model Building

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
# initialize the model
ann = Sequential()

# Add hidden layers
ann.add(Dense(units = 6, activation='relu'))

# Add output layer
ann.add(Dense(units = 1, activation='sigmoid'))

# Establishing the relationship between layer
ann.compile(optimizer='adam', loss = 'binary_crossentropy')

ann.fit(xtrain,ytrain,epochs=50)
```

```python
yp = ann.predict(xtest)

yp = yp > 0.5
yp
```

```python
print(classification_report(ytest,yp))
```

```python
ann = Sequential()

ann.add(Dense(units=30,activation='relu'))
ann.add(Dense(units=25,activation='relu'))

ann.add(Dense(units=1,activation='tanh'))

ann.compile(optimizer='adam',loss='binary_crossentropy')

ann.fit(xtrain,ytrain,validation_data=(xtest,ytest),epochs=70)
```

```python
val_loss = pd.DataFrame(ann.history.history)
val_loss.plot()
```
